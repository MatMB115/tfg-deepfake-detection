### Unzip Train Set

```python
import zipfile
import os

zip_file_path = 'DeepfakeBench/datasets/rgb/train.zip'

extract_dir = 'DeepfakeBench/datasets/rgb/'

os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

os.remove(zip_file_path)

print(f'Arquivos extraídos para {extract_dir} e o arquivo zip foi deletado.')
```
